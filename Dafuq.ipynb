```python
import pandas as pd  # Mengimpor pustaka pandas untuk manipulasi data.
import matplotlib.pyplot as plt  # Mengimpor pustaka matplotlib untuk visualisasi data.
import seaborn as sns  # Mengimpor pustaka seaborn untuk visualisasi data statistik.
import nltk  # Mengimpor pustaka nltk untuk pemrosesan bahasa alami.
from nltk.sentiment import SentimentIntensityAnalyzer  # Mengimpor kelas untuk analisis sentimen.
import networkx as nx  # Mengimpor pustaka networkx untuk analisis jaringan.
```

```python
# Memuat Data
# Tujuan: Membaca data dari file CSV ke DataFrame pandas.
```

```python
# 2. Memuat Data
# Tujuan: Membaca data dari file CSV ke DataFrame pandas.
df_fantasi = pd.read_csv('kesatu.csv')  # Membaca data fantasi dari kesatu.csv.
df_hobi = pd.read_csv('kedua.csv')  # Membaca data hobi dari kedua.csv.
df_julukan = pd.read_csv('ketiga.csv')  # Membaca data julukan dari ketiga.csv.
df_ibu_ayah = pd.read_csv('keempat.csv')  # Membaca data ibu ayah dari keempat.csv.
df_evolusi = pd.read_csv('kelima.csv')  # Membaca data evolusi dari kelima.csv.
```

```python
# Membersihkan Data 
# Tujuan: Menghapus baris dengan nilai NaN.
```

```python
df_fantasi.dropna(inplace=True)  # Menghapus baris NaN di data fantasi.
df_hobi.dropna(inplace=True)  # Menghapus baris NaN di data hobi.
df_julukan.dropna(inplace=True)  # Menghapus baris NaN di data julukan.
df_ibu_ayah.dropna(inplace=True)  # Menghapus baris NaN di data ibu ayah.
df_evolusi.dropna(inplace=True)  # Menghapus baris NaN di data evolusi.

```

```python
# Menampilkan Informasi Data
# Tujuan: Menampilkan ringkasan DataFrame.
```

```python
print("Info Data Fantasi Lokal:")  # Menampilkan info data fantasi.
print(df_fantasi.info())  # Menampilkan info data fantasi.
print("\nInfo Data Hobi:")  # Menampilkan info data hobi.
print(df_hobi.info())  # Menampilkan info data hobi.
print("\nInfo Data Julukan:")  # Menampilkan info data julukan.
print(df_julukan.info())  # Menampilkan info data julukan.
print("\nInfo Data Ibu Ayah:")  # Menampilkan info data ibu ayah.
print(df_ibu_ayah.info())  # Menampilkan info data ibu ayah.
print("\nInfo Data Evolusi:")  # Menampilkan info data evolusi.
print(df_evolusi.info())  # Menampilkan info data evolusi.
```

```python
# Visualisasi Distribusi Ras/Spesies dan Kerajaan
# Tujuan: Menampilkan distribusi ras/spesies dan kerajaan.
```

```python
plt.figure(figsize=(8, 8))  # Membuat figure untuk pie chart.
plt.pie(df_fantasi['Ras/Spesies'].value_counts(), labels=df_fantasi['Ras/Spesies'].unique(), autopct='%1.1f%%')  # Membuat pie chart.
plt.title('Distribusi Ras/Spesies Karakter Fantasi Lokal')  # Menambahkan judul.
plt.show()  # Menampilkan pie chart.
pd.crosstab(df_fantasi['Kerajaan'], df_fantasi['Ras/Spesies']).plot(kind='bar', stacked=True, figsize=(10, 6))  # Membuat stacked bar chart.
plt.title('Distribusi Ras/Spesies dalam Setiap Kerajaan')  # Menambahkan judul.
plt.show()  # Menampilkan stacked bar chart.



```

```python
# Visualisasi Distribusi Tingkat Kekuatan
# Tujuan: Menampilkan distribusi tingkat kekuatan karakter.
```

```python
if 'Tingkat Kekuatan' in df_fantasi.columns:  # Memeriksa keberadaan kolom 'Tingkat Kekuatan'.
    plt.figure(figsize=(10, 6))  # Membuat figure untuk countplot.
    sns.countplot(x='Tingkat Kekuatan', data=df_fantasi)  # Membuat countplot.
    plt.title('Distribusi Tingkat Kekuatan Karakter')  # Menambahkan judul.
    plt.show()  # Menampilkan countplot.
```

```python
# Visualisasi Distribusi Hobi dan Kesukaan
# Tujuan: Menampilkan frekuensi hobi karakter.
```

```python
plt.figure(figsize=(10, 6))  # Membuat figure untuk countplot.
sns.countplot(y='Hobi/Kesukaan', data=df_hobi, order=df_hobi['Hobi/Kesukaan'].value_counts().index)  # Membuat countplot.
plt.title('Frekuensi Hobi Karakter Fantasi Lokal')  # Menambahkan judul.
plt.show()  # Menampilkan countplot.

```

```python
# Visualisasi Kill Rate dan Hubungan dengan Julukan
# Tujuan: Menampilkan kill rate dan hubungannya dengan julukan.
```

```python
plt.figure(figsize=(10, 6))  # Membuat figure untuk bar chart.
plt.bar(df_julukan['Julukan'], df_julukan['Kill Rate (Perkiraan)'])  # Membuat bar chart.
plt.xticks(rotation=45)  # Memutar label sumbu x.
plt.title('Kill Rate Karakter Fantasi Lokal')  # Menambahkan judul.
plt.xlabel('Julukan')  # Menambahkan label sumbu x.
plt.ylabel('Kill Rate (Perkiraan)')  # Menambahkan label sumbu y.
plt.show()  # Menampilkan bar chart.


if 'Kill Rate (Perkiraan)' in df_julukan.columns:  # Memeriksa keberadaan kolom 'Kill Rate'.
    plt.figure(figsize=(10, 6))  # Membuat figure untuk scatter plot.
    plt.scatter(df_julukan['Kill Rate (Perkiraan)'], range(len(df_julukan)))  # Membuat scatter plot.
    plt.yticks(range(len(df_julukan)), df_julukan['Julukan'])  # Mengatur label sumbu y.
    plt.title('Hubungan antara Julukan dan Kill Rate (Perkiraan)')  # Menambahkan judul.
    plt.xlabel('Kill Rate (Perkiraan)')  # Menambahkan label sumbu x.
    plt.ylabel('Julukan')  # Menambahkan label sumbu y.
    plt.show()  # Menampilkan scatter plot.

if 'Kill Rate' in df_julukan.columns:  # Memeriksa keberadaan kolom 'Kill Rate'.
    plt.figure(figsize=(8, 6))  # Membuat figure untuk heatmap.
    sns.heatmap(df_julukan[['Kill Rate (Perkiraan)']].corr(), annot=True, cmap='coolwarm')  # Membuat heatmap.
    plt.title('Heatmap Korelasi Kill Rate (Perkiraan)')  # Menambahkan judul.
    plt.show()  # Menampilkan heatmap.

```

```python
# Visualisasi Distribusi Potensi Evolusi
```

```python
import matplotlib.pyplot as plt # Mengimpor library matplotlib.pyplot untuk membuat visualisasi grafik.

# Pie Chart (Diagram Lingkaran)
plt.figure(figsize=(8, 8)) # Membuat objek gambar dengan ukuran 8x8 inci.
plt.pie(df_evolusi['Potensi Evolusi'].value_counts(), # Membuat diagram lingkaran dari jumlah kemunculan setiap nilai di kolom 'Potensi Evolusi'.
        labels=df_evolusi['Potensi Evolusi'].unique(), # Memberikan label pada setiap bagian lingkaran dengan nilai unik dari kolom 'Potensi Evolusi'.
        autopct='%1.1f%%', # Menampilkan persentase setiap bagian lingkaran dengan satu angka desimal.
        wedgeprops=dict(width=0.4)) # Mengatur lebar bagian lingkaran untuk membuat "donat chart".
plt.title('Distribusi Potensi Evolusi Karakter') # Menambahkan judul pada diagram lingkaran.
plt.show() # Menampilkan diagram lingkaran.

# Bar Chart (Diagram Batang)
plt.figure(figsize=(10, 6)) # Membuat objek gambar dengan ukuran 10x6 inci.
plt.bar(df_evolusi['Potensi Evolusi'].value_counts().index, # Membuat diagram batang dengan indeks dari jumlah kemunculan sebagai sumbu x.
        df_evolusi['Potensi Evolusi'].value_counts().values) # Menggunakan nilai jumlah kemunculan sebagai tinggi batang.
plt.xticks(rotation=45, ha='right') # Memutar label sumbu x sebesar 45 derajat agar mudah dibaca.
plt.title('Frekuensi Potensi Evolusi Karakter Fantasi Lokal') # Menambahkan judul pada diagram batang.
plt.xlabel('Potensi Evolusi') # Memberikan label pada sumbu x.
plt.ylabel('Frekuensi') # Memberikan label pada sumbu y.
plt.tight_layout() # Mengatur tata letak agar semua elemen diagram batang terlihat rapi.
plt.show() # Menampilkan diagram batang.
```

```python
# Analisis Sentimen Julukan
```

```python
sia = SentimentIntensityAnalyzer()
df_julukan['Sentimen'] = df_julukan['Julukan'].apply(lambda text: sia.polarity_scores(text)['compound'])

plt.figure(figsize=(10, 6))
plt.hist(df_julukan['Sentimen'], bins=20)
plt.title('Distribusi Sentimen Julukan')
plt.xlabel('Sentimen')
plt.ylabel('Frekuensi')
plt.show()

if 'Kill Rate' in df_julukan.columns:
    plt.figure(figsize=(10, 6))
    plt.scatter(df_julukan['Sentimen'], df_julukan['Kill Rate (Perkiraan)'])
    plt.title('Sentimen Julukan vs. Kill Rate (Perkiraan)')
    plt.xlabel('Sentimen')
    plt.ylabel('Kill Rate (Perkiraan)')
    plt.show()
```

```python
# 12. Analisis Jaringan Pasangan Ibu dan Ayah
```

```python
import networkx as nx # Mengimpor library networkx untuk membuat grafik.
import matplotlib.pyplot as plt # Mengimpor library matplotlib.pyplot untuk membuat visualisasi grafik.

G = nx.Graph() # Membuat objek grafik kosong.

for _, row in df_ibu_ayah.iterrows(): # Mengiterasi setiap baris dalam DataFrame df_ibu_ayah.
    G.add_edge(row['Nama Ibu,Nama Ayah'], row['Nama Ibu,Nama Ayah']) # Menambahkan edge (hubungan) antara 'Nama Ibu' dan 'Nama Ayah' ke grafik.

plt.figure(figsize=(12, 12)) # Membuat objek gambar dengan ukuran 12x12 inci.
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1500, edge_color='gray') # Menggambar grafik dengan label node, warna, ukuran, dan warna edge.
plt.title('Jaringan Pasangan Ibu dan Ayah') # Menambahkan judul pada gambar.
plt.show() # Menampilkan gambar grafik.
```
